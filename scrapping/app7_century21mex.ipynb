```python
def depabye(x):
    x=x.upper().replace(",","")
    if len(x.split("INT"))>1:
        x=x.split("INT")[0]
    elif len(x.split("DEPTO"))>1:
        x=x.split("DEPTO")[0]
    elif len(x.split("-"))>1:
        x=x.split("-")[0]
    
    return(x)
    
```

```python
l=[]
base_url="https://century21mexico.com/departamento/venta/Ciudad-De-Mexico/Benito-Ju%C3%A1rez?pagina="
base="https://century21mexico.com/departamento/venta/Ciudad-De-Mexico/Benito-Ju%C3%A1rez"
#page=1
for pg in range(page_nr):
    pg=pg+1
    r=requests.get(base_url+str(pg)+".html")
    c=r.content
    soup=BeautifulSoup(c,"html.parser")
    all=soup.find_all("div",{"col-md-6 col-lg-4 mb-3"})
    for item in all:
        d={}
        d['direccion']=str(item.find_all("a")).replace("\n","").split("/")[6].replace("'","").split(">")[0]
        d['Precio']=float(item.find("h5",{"class":"h6 card-title precio family-typoldBold rGold"}).text.replace("\n","").replace(" ","").replace("$","").replace(",","").split("MXN")[0]
         )
        d['colonia']=item.find("span",{"class":"family-oakesReg"}).text.split("\n")[0]
        d['alcaldia']=item.find("span",{"class":"family-oakesReg"}).text.split("\n")[1]
        d['ciudad']=item.find("span",{"class":"family-oakesReg"}).text.split("\n")[2]
        d['Tels']=item.find("span",{"class":"font-weight-bold family-typoldBold"}).text
        d['n_recamaras']=int(item.find("div",{"class":"col-4 typoldReg text-right"}).text.replace("\n","").replace(" ","").split("\xa0")[0])
        d['n_banios']=int(item.find("div",{"class":"col-4 typoldReg text-right"}).text.replace("\n","").replace(" ","").split("\xa0")[1])
        d['n_estacionam']=int(item.find("div",{"class":"col-4 typoldReg text-right"}).text.replace("\n","").replace(" ","").split("\xa0")[2])
        d['Terreno']=int(item.find("div",{"class":"col-4 typoldReg text-right"}).text.replace("\n","").replace(" ","").split("\xa0")[3])
        ent1=item.find("span",{"class":"family-oakesReg"}).text.split("\n")[0]
        entra=str(item.find_all("a")).replace("\n","").split("/")[6].replace("'","").split(">")[0]
        qwr=requests.get(base+"/"+str(ent1).replace(" ","-")+"/"+entra.split('"')[0]+".html")
        cw=qwr.content
        soup2=BeautifulSoup(cw,"html.parser")
        all2=soup2.find_all("span",{"itemprop":"streetAddress"})
        try:
            d['calle']=str(all2[0].text)
        except:
            d['calle']=None

        
        try:
            d['construccion']=int(item.find("div",{"class":"col-4 typoldReg text-right"}).text.replace("\n","").replace(" ","").split("\xa0")[4])
        except:
            d['construccion']=None

        l.append(d)
```

```python
l=[]
base_url="https://www.pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s="
for page in range(0,int(page_nr)*10,10):
    print( )
    r=requests.get(base_url+str(page)+".html")
    c=r.content
    #c=r.json()["list"]
    soup=BeautifulSoup(c,"html.parser")
    all=soup.find_all("div",{"class":"propertyRow"})
    for item in all:
        d={}
        d["Address"]=item.find_all("span",{"class","propAddressCollapse"})[0].text
        try:
            d["Locality"]=item.find_all("span",{"class","propAddressCollapse"})[1].text
        except:
            d["Locality"]=None
        d["Price"]=item.find("h4",{"class","propPrice"}).text.replace("\n","").replace(" ","")
        try:
            d["Beds"]=item.find("span",{"class","infoBed"}).find("b").text
        except:
            d["Beds"]=None
    
        try:
            d["Area"]=item.find("span",{"class","infoSqFt"}).find("b").text
        except:
            d["Area"]=None
    
        try:
            d["Full Baths"]=item.find("span",{"class","infoValueFullBath"}).find("b").text
        except:
            d["Full Baths"]=None

        try:
            d["Half Baths"]=item.find("span",{"class","infoValueHalfBath"}).find("b").text
        except:
            d["Half Baths"]=None
        for column_group in item.find_all("div",{"class":"columnGroup"}):
            for feature_group, feature_name in zip(column_group.find_all("span",{"class":"featureGroup"}),column_group.find_all("span",{"class":"featureName"})):
                if "Lot Size" in feature_group.text:
                    d["Lot Size"]=feature_name.text
        l.append(d)
```

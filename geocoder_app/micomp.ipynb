```python
import numpy as np
import pandas as pd
import folium
from geopy.geocoders import Nominatim
from geopy import distance,Point
```

```python
df=pd.read_csv('dc.csv')
#df[df['NB_COM']=="GASOL BP FELIX CU A8"]
```

```python
df=df.drop("index",axis=1)
df=df.groupby(['NB_COM','CP'],as_index=False).count()
#df.sort_values(by="IM_TR",ascending=False)
df[df['IM_TR']==6]
```

```python
df=df.sort_values(by="IM_TR",ascending=True)
#str(df['CP'].iloc[7])[-5:]
df[df['IM_TR']==6]
```

```python
df['CP']=df["CP"].apply(lambda x: str(x)[-5:])
df
```

```python
nom=Nominatim(timeout=10)
df['dir']=df['CP']+","+"mexico"
df.head()
```

```python
#df['dir']=df['dir'].apply(lambda x: x.replace(" ,",",").replace(", ",","))
#geo=df['dir'].apply(lambda x: nom.geocode(x))df['dir']=df['dir'].apply(lambda x: x.replace(" ,",",").replace(", ",","))
#geo=df['dir'][0:10].apply(lambda x: nom.geocode(x))
##df['CP2']=df['CP'].apply(lambda x: str(x)[1:])
#df['dir2']=df['CP2']+","+"mexico"
#df.head()
nom.geocode(df['dir'].iloc[20])
```

```python
geo=df['dir'].apply(lambda x: nom.geocode(x))
```

```python
lat=list([x.latitude for x in geo if x!=None])
long=list([x.longitude for x in geo if x!=None])
ind=np.where(np.array(geo)!=None)
cost=df['NB_COM'].iloc[ind]
trx=df['IM_TR'].iloc[ind]
```

```python
map=folium.Map(location=[lat[0], long[0]], zoom_start=13)
fg=folium.FeatureGroup(name="Comercio")
for lt, ln,ct,tr in zip(lat, long,cost,trx):#, elev, name):
    #iframe = folium.IFrame(html=html % (name, name, el), width=200, height=100)
    fg.add_child(folium.Marker(location=[lt, ln],popup="Comercio:"+str(ct)+" "+"n_trax:"+str(tr), icon = folium.Icon(color = "green")))
#fg.add_child(folium.Marker(location=[lat[1], long[1]],popup="cp",icon=folium.Icon(color='red'))) 
                                     
#fg.add_child(folium.Marker(location=[ma.latitude, ma.longitude],popup="ma",icon=folium.Icon(color='red')))
#fg.add_child(folium.Marker(location=[sc.latitude, sc.longitude],popup="sc",icon=folium.Icon(color='red')))
map.add_child(fg)
```
